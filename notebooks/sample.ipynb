```python
df["Family_members_on_ship"] = df["SibSp"] + df["Parch"]

df = df.drop(["SibSp","Parch"],axis=1)
```

```python
df["Family_on_board"] = df["Family_members_on_ship"] > 0
df
```

```python
df_numerical = df[["Age","Fare","Family_members_on_ship"]]
df_categorical = df[["Survived","Pclass","Sex","Cabin","Embarked","Family_on_board"]]
```

```python
def numerical_stats(num):
    stats = {}
    for i in num:
        stats[i] = { 
            "mean": num[i].mean(),
            "median": num[i].median(),
            "min": num[i].min(),
            "max": num[i].max(),
            "std": num[i].std(),
            "missing": num[i].isnull().sum()
        }
    return stats
```

```python
def most_feq(num):
    max_count = num.value_counts().max()
    mask = num.value_counts() == max_count
    return num.value_counts()[mask].index.to_numpy()
```

```python
def categorical_stats(num):
    stats = {}
    for i in num:
        stats[i] = {
             "most_frequent_value": most_feq(num[i]),
             "unique_counts": num[i].nunique()
        }
    return stats
```

```python
def stats(numerical_df,categorical_df):
    stats = {
        "numerical_col": numerical_stats(numerical_df),
        "categorical_col": categorical_stats(categorical_df)
    }

    return stats

```

```python
stats(df_numerical,df_categorical)
```
